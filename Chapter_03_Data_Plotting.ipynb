# Chapter 3: Data Visualization & Plotting

Data visualization is key to exploratory data analysis (EDA). Before building a machine learning model, you must use your own eyes to study the distribution, trends, correlation, and anomalies in your data.

### High-Level Intuition & Analogy
If you look at a spreadsheet of 100,000 house prices, you won't see anything but numbers. But if you plot a **scatter plot** of house size vs. price, you will immediately see if the relationship is linear, curved, or if there are weird outliers (like a tiny house selling for $10 million). 
* **Matplotlib** is your drawing board - it gives you low-level controls over every line, tick mark, and text label.
* **Seaborn** is your high-level artist - it takes DataFrames directly and creates gorgeous, statistically rich plots automatically.

### Core Concepts & Step-by-Step Execution
1. **Histograms**: Plots the distribution of a single numerical column by grouping values into bins, showing you if the data is symmetric or skewed.
2. **Scatter Plots with Regression**: Shows the relationship between two variables, highlighting how they correlate. Adding a trendline highlights the general direction.
3. **Correlation Heatmaps**: A grid showing the correlation coefficient ($r$) between every pair of features in the dataset. We mask the upper diagonal to avoid redundant information.

### Mathematical Foundations (Simplified)
* **Pearson Correlation Coefficient ($r$)**: Measures the linear relationship between two variables, ranging from -1 (perfect negative correlation) to +1 (perfect positive correlation), with 0 representing no correlation.
  $$r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}}$$
* **Anscombe's Quartet**: A set of four datasets that have identical simple statistical properties (mean, variance, correlation) but display completely different distributions when graphed. This highlights the absolute necessity of visualization.

### Pros and Cons
* **Pros**: Intuitive patterns identification, simple outliers spotting, and easy communication of results.
* **Cons**: Over-plotting can hide data patterns in large datasets, and static charts cannot show multi-dimensional interactions.
* **Interactive Alternatives**: For interactive web dashboards, libraries like **Plotly** or **Bokeh** are preferred over static Seaborn plots.

### Pro-Tips & Gotchas
* **Misleading Scales**: Always check your axis scales! Starting y-axes at values other than zero can exaggerate small differences, leading to wrong conclusions.

### Programming Guide & Key Functions
Here is a comprehensive visual guide demonstrating core visualization functions:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Creating sample data
df = pd.DataFrame({
    'Age': [23, 25, 30, 35, 40, 45, 50, 55, 60],
    'Salary': [45000, 48000, 54000, 65000, 70000, 80000, 85000, 95000, 110000],
    'Purchased': ['No', 'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes']
})

# --- Matplotlib Plotting Structure ---
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # Create subplots (1 row, 2 columns)

# Subplot 1: Line plot
axes[0].plot(df['Age'], df['Salary'], marker='o', color='blue', linestyle='--')
axes[0].set_title('Salary Trend by Age')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Salary')
axes[0].grid(True)

# Subplot 2: Histogram of distribution
axes[1].hist(df['Salary'], bins=3, color='orange', edgecolor='black')
axes[1].set_title('Salary Distribution')
axes[1].set_xlabel('Salary Range')

plt.tight_layout()  # Adjust layouts to avoid overlapping text
plt.show()

# --- Seaborn Statistical Plots ---
# Scatter plot with regression line
sns.lmplot(x='Age', y='Salary', hue='Purchased', data=df)
plt.title('Salary vs Age with Regression Line')
plt.show()

# Box Plot & Violin Plot (distribution comparisons)
sns.boxplot(x='Purchased', y='Salary', data=df)  # Box plot
plt.show()
sns.violinplot(x='Purchased', y='Salary', data=df)  # Violin plot (adds kernel density estimates)
plt.show()

# Correlation Heatmap
corr = df.corr(numeric_only=True)
mask = np.triu(np.ones_like(corr, dtype=bool))  # Create mask for upper diagonal
sns.heatmap(corr, mask=mask, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Masked Correlation Heatmap')
plt.show()
```
