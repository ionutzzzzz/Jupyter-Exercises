# Chapter 5: Decision Trees & Random Forests

Decision Trees and Random Forests are intuitive, rule-based algorithms used for classification and regression tasks.

### High-Level Intuition & Analogy
* **Decision Tree** is like playing a game of "20 Questions." You start at the root and ask a simple question (e.g. "Is age > 30?"). Depending on the answer, you go left or right and ask another question, until you reach a leaf node (the prediction).
* **Random Forest** is the wisdom of the crowd. Instead of trusting a single decision tree (which might be biased or memorize details), we train a forest of 100 trees on random subsets of the data and features. To get a final prediction, we let all 100 trees vote and take the majority decision.

### Core Concepts & Step-by-Step Execution
1. **Tree Growing**: The tree searches all features for the split that creates the cleanest resulting groups (highest purity).
2. **Impurity Metric**: Splitting decisions are driven by metrics like **Gini Impurity** or **Entropy**.
3. **Ensemble Aggregation**: Random Forests run bootstrap sampling (training each tree on a random slice of rows) and feature bagging (selecting random features for each split) to ensure trees are independent.
4. **Out-of-Bag (OOB) Score**: A built-in validation score for Random Forests calculated using samples that were left out during the bootstrap training of each tree.

### Mathematical Foundations (Simplified)
* **Shannon Entropy**: Measures the amount of uncertainty or impurity in a dataset.
  $$H(S) = -\sum p_i \log_2(p_i)$$
  Where $p_i$ is the probability of class $i$. Information Gain is the reduction in entropy achieved by partitioning the data.
* **Gini Impurity**: Measures how often a randomly chosen element would be incorrectly labeled.
  $$G = 1 - \sum (p_i)^2$$

### Pros and Cons
* **Pros**: No feature scaling required, handles non-linear relationships naturally, and offers excellent interpretability (for single trees).
* **Cons**: Single trees overfit very easily, and large random forests act as "black box" models.
* **Comparison**: Linear models work best on linear datasets; tree-based models perform significantly better on tabular data with complex interactions.

### Pro-Tips & Gotchas
* **Tree Pruning**: To prevent overfitting, always limit the depth of your decision trees using hyperparameters like `max_depth`, `min_samples_split`, or `min_samples_leaf`.

### Programming Guide & Key Functions
Here is a comprehensive guide showing tree visualization and ensemble configuration:

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier

# Sample classification splits
X = [[25, 50000], [35, 70000], [45, 90000], [20, 30000], [55, 120000]]
y = [0, 0, 1, 0, 1]

# --- Training and Visualising a Decision Tree ---
dt = DecisionTreeClassifier(max_depth=3, min_samples_split=2, criterion='gini', random_state=42)
dt.fit(X, y)

# Visualise the decision tree structure
plt.figure(figsize=(10, 8))
plot_tree(dt, feature_names=['Age', 'Salary'], class_names=['No', 'Yes'], filled=True)
plt.show()

# --- Random Forest Ensemble Configuration ---
rf = RandomForestClassifier(
    n_estimators=100,      # Number of decision trees
    max_depth=5,           # Constraint on tree depth
    max_features='sqrt',   # Pick random features for each split
    bootstrap=True,        # Use bootstrap sampling
    oob_score=True,        # Calculate Out-of-Bag validation score
    random_state=42
)
rf.fit(X, y)

print("OOB validation score:", rf.oob_score_)
print("Feature importances:", rf.feature_importances_)  # [Importance_Age, Importance_Salary]
```

````python
import numpy as np

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Demonstration
iris = load_iris()
X_tr, X_te, y_tr, y_te = train_test_split(
    iris.data, iris.target, test_size=0.2, random_state=42
)
dt = DecisionTreeClassifier(max_depth=2, random_state=42).fit(X_tr, y_tr)
print("Decision Tree Train Acc:", dt.score(X_tr, y_tr))
print("Decision Tree Test Acc:", dt.score(X_te, y_te))
````
