### Exercise 6: Word Importance via Model Coefficients (Hard)
Because Logistic Regression is a linear model, the magnitude and sign of its coefficients directly represent word importance (positive coefficients indicate positive sentiment correlation, negative indicate negative correlation).
Write a function `get_top_positive_words(model, feature_names, top_n=5)` that:
1. Extracts the coefficients array of the model (use `model.coef_[0]`).
2. Finds the indices of the `top_n` largest coefficient values (most positive).
3. Maps these indices to their corresponding words in the `feature_names` list.
4. Returns a list of strings of the top words, sorted from largest coefficient to smallest.


```python
def get_top_positive_words(model, feature_names, top_n=5):
    # TODO: Extract coefficients and return top_n words predicting positive sentiment
    coefs = model.coef_[0]
    top_indices = np.argsort(coefs)[::-1][:top_n]
    return [feature_names[i] for i in top_indices]
```
