# Chapter 17: Association Rule Learning

Association Rule Learning discovers interesting relationships between items in large transaction databases.

### High-Level Intuition & Analogy
This is the "market basket analysis" technique (e.g. discovering that people who buy diapers also frequently buy beer).

### Core Concepts & Step-by-Step Execution
1. **Frequent Itemset Generation**: Apriori identifies items that appear above a minimum support threshold.
2. **Rule Generation**: Extract rules from frequent itemsets.
3. **Metric Pruning**: Filter rules based on confidence and lift.

### Mathematical Foundations (Simplified)
1. **Support**: $\text{Support}(A) = \frac{\text{Transactions containing } A}{\text{Total Transactions}}$
2. **Confidence**: $\text{Confidence}(A \rightarrow B) = \frac{\text{Support}(A \cup B)}{\text{Support}(A)}$
3. **Lift**: $\text{Lift}(A \rightarrow B) = \frac{\text{Confidence}(A \rightarrow B)}{\text{Support}(B)}$

### Pros and Cons
* **Pros**: Easy to interpret, and helps with store layout optimization.
* **Cons**: Can generate thousands of redundant rules, requiring strict pruning.

### Pro-Tips & Gotchas
* **High Lift / Low Support**: Beware of rules with high lift but tiny support. They might represent rare combinations that occurred by pure chance. Always set a sensible minimum support threshold.

### Programming Guide & Key Functions
Here is a comprehensive script using Pandas to calculate metrics:

```python
import pandas as pd

# Transaction database representing items purchased
df = pd.DataFrame({
    'Bread': [1, 1, 0, 1, 0],
    'Milk':  [1, 0, 1, 1, 1],
    'Beer':  [0, 1, 0, 1, 1]
})

# --- Support Calculation ---
support_bread = df['Bread'].mean()
support_milk = df['Milk'].mean()
support_bread_milk = (df['Bread'] & df['Milk']).mean()

# --- Rule metrics Bread -> Milk ---
confidence = support_bread_milk / support_bread
lift = confidence / support_milk

print(f"Bread support: {support_bread:.2f}")
print(f"Bread -> Milk Confidence: {confidence:.2f}")
print(f"Rule Lift score: {lift:.2f}")
```
