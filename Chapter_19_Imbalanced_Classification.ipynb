# Chapter 19: Classification on Imbalanced Data

Class imbalance occurs when the classes in a classification dataset are not represented equally. For example, in fraud detection, only 0.1% of transactions may be fraudulent.

## 1. The Accuracy Metric Trap
When classes are highly imbalanced, **accuracy** is a misleading metric. If 99% of samples are of class 0, a naive model predicting 0 for all samples achieves 99% accuracy but fails completely. We must evaluate models using:
- **Precision**: $\frac{TP}{TP + FP}$ (Fraction of predicted positives that are correct).
- **Recall**: $\frac{TP}{TP + FN}$ (Fraction of actual positives found).
- **F1 Score**: $2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}$ (Harmonic mean).
- **Macro F1 Score**: Averaging the F1 scores calculated separately for each class.

## 2. Resampling Techniques
- **Oversampling**: Duplicating minority class samples or synthesizing new ones (**SMOTE**).
- **Undersampling**: Removing majority class samples to balance frequencies.

## 3. Cost-Sensitive Learning
Instead of balancing the datasets, we modify the loss function during training, assigning a higher weight (penalty) to misclassifications of the minority class. In scikit-learn, this is set using the parameter `class_weight='balanced'`.


## 🛠️ Key Functions to Keep in Mind
* **Data Inspection**:
  * `np.unique(y, return_counts=True)`: Identifies class labels and counts occurrences (reveals imbalance).
* **Scikit-Learn Classification**:
  * `SVC(class_weight='balanced')`: SVM classifier penalizing minority errors inversely proportional to class frequencies.
  * `DecisionTreeClassifier(class_weight='balanced')`: Decision tree adjusting split selection calculations for class weights.
* **Evaluation**:
  * `balanced_accuracy_score(y_true, y_pred)`: Computes balanced accuracy (average recall obtained on each class).



```python

from sklearn.datasets import make_classification
from sklearn.metrics import classification_report
from sklearn.svm import SVC
import numpy as np

# Highly imbalanced dataset
X, y = make_classification(n_samples=1000, weights=[0.95, 0.05], random_state=42)
print("Class distribution:", np.bincount(y))

# Train classifier
clf = SVC(kernel='linear', random_state=42).fit(X, y)
print("Classification Report:")
print(classification_report(y, clf.predict(X)))


```
