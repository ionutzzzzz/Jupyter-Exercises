# Chapter 22: Deep Computer Vision & Convolutional Networks

Deep Computer Vision leverages Convolutional Neural Networks (CNNs) to automatically extract hierarchical spatial features from grid-structured data like images.

## 1. Image Representation
An image is represented as a 3D NumPy array of shape $(H, W, C)$:
- $H$: Height in pixels.
- $W$: Width in pixels.
- $C$: Channels (e.g., 3 for RGB, 1 for Grayscale).
Pixel values are typically integers in $[0, 255]$ or scaled floats in $[0, 1]$.

## 2. The 2D Convolution Operation
Convolution slides a small kernel matrix $K$ (e.g. $3 \times 3$) over the image $I$ to extract features like edges or textures. At each position:
$$O(i, j) = \sum_{m} \sum_{n} I(i + m, j + n) \cdot K(m, n)$$

## 3. Max Pooling
Pooling downsamples feature maps to reduce computational cost, memory, and to build translation invariance. Max Pooling partitions the grid into blocks (e.g. $2 \times 2$) and outputs the maximum value from each block.

## 4. Activation Functions
The Rectified Linear Unit (**ReLU**) is standard in CNNs to introduce non-linearity:
$$f(x) = \max(0, x)$$

## 🛠️ Key Functions to Keep in Mind
* **PIL (Pillow) & SciPy convolve2d**:
  * `Image.open(img_path)`: Opens an image.
  * `img.convert('L')`: Converts to grayscale.
  * `scipy.signal.convolve2d(in1, in2, mode)`: Helper for fast 2D convolution.
* **NumPy Functions**:
  * `np.pad(array, pad_width, mode)`: Pads arrays (useful for padding pixels).
  * `np.max(array)`: Maximum element.


```python
import numpy as np
from scipy.signal import convolve2d

# Example 2D convolution
img_dummy = np.array([
    [1, 1, 1],
    [0, 0, 0],
    [1, 1, 1]
])
kernel = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
])
res = convolve2d(img_dummy, kernel, mode='valid')
print("Convolved Feature Map:\n", res)
```
