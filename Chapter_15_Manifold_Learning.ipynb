# Chapter 15: Dimensionality Reduction & Manifold Learning

This unit covers methods for projecting high-dimensional data into lower dimensions.

### High-Level Intuition & Analogy
* **LDA (Linear Discriminant Analysis)** is like PCA, but it knows about classes. It finds the projection that keeps different groups as far apart as possible.
* **t-SNE** is a visualization tool that maps high-dimensional points to 2D. It preserves local neighborhoods, making it easy to see clusters visually.

### Core Concepts & Step-by-Step Execution
1. **LDA Projection**: Identifies the axes that maximize difference between class means while minimizing variance within classes.
2. **t-SNE Minimization**: Converts high-dimensional and low-dimensional distances into probabilities, and minimizes difference using gradient descent.
3. **Silhouette Evaluation**: Computes distances between cluster members vs. neighboring clusters.

### Mathematical Foundations (Simplified)
* **LDA Objective**: Maximizes ratio of between-class scatter matrix $S_B$ to within-class scatter matrix $S_W$.
  $$J(\mathbf{w}) = \frac{\mathbf{w}^T S_B \mathbf{w}}{\mathbf{w}^T S_W \mathbf{w}}$$
* **Silhouette Score**: For a point $i$: $s(i) = \frac{b(i) - a(i)}{\max(a(i), b(i))}$, where $a(i)$ is mean in-cluster distance and $b(i)$ is mean distance to the nearest neighboring cluster.

### Pros and Cons
* **Pros**: Improves visualization, and prevents overfitting by removing redundant dimensions.
* **Cons**: Non-linear manifold methods are slow, and t-SNE scales poorly to large datasets.

### Pro-Tips & Gotchas
* **t-SNE Perplexity**: The `perplexity` parameter in t-SNE balances attention between local and global data structures. Setting it too low will split single clusters into multiple tiny ones.

### Programming Guide & Key Functions
Here is a comprehensive dimensionality reduction and evaluation pipeline:

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.manifold import TSNE
from sklearn.metrics import silhouette_score, silhouette_samples
import numpy as np

X = np.random.rand(100, 10)
y = np.random.randint(0, 3, 100)

# --- Linear Discriminant Analysis (Supervised) ---
lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X, y)

# --- t-SNE Projection (Non-linear) ---
tsne = TSNE(
    n_components=2, 
    perplexity=15.0,  # Balance local and global attention
    n_iter=1000, 
    random_state=42
)
X_tsne = tsne.fit_transform(X)

# --- Silhouette Evaluation Metrics ---
score_avg = silhouette_score(X_tsne, y)
print("Average silhouette score:", score_avg)

# Get sample-specific silhouette scores
sample_scores = silhouette_samples(X_tsne, y)
```

````python
import numpy as np

from sklearn.datasets import make_circles
from sklearn.decomposition import KernelPCA
import matplotlib.pyplot as plt

# Generate non-linear concentric circles
X, y = make_circles(n_samples=100, factor=0.5, noise=0.05, random_state=42)

# Fit Kernel PCA
kpca = KernelPCA(n_components=2, kernel="rbf", gamma=15)
X_kpca = kpca.fit_transform(X)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap="viridis")
axes[0].set_title("Original Data Space")
axes[1].scatter(X_kpca[:, 0], X_kpca[:, 1], c=y, cmap="viridis")
axes[1].set_title("Kernel PCA Space")
plt.show()
````
