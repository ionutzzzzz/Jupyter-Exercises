### Exercise 5: Preprocess Titanic Dataset (Medium)
To practice classification on a different tabular dataset, let's load `data/titanic.csv`.
Write a function `preprocess_titanic(file_path)` that:
1. Loads the Titanic dataset.
2. Imputes missing values in `Age` with the median age.
3. Encodes the categorical column `Sex` using map: `{'male': 0, 'female': 1}`.
4. Drops columns `PassengerId` and `Embarked` (along with the target `Survived` from the features).
5. Returns `(X, y)` where `X` is the features DataFrame and `y` is the target array (`Survived`).


```python
def preprocess_titanic(file_path):
    # TODO: Load csv, fillna Age with median, map Sex, drop PassengerId and Embarked, return X and y
    df = pd.read_csv(file_path)
    df["Age"] = df["Age"].fillna(df["Age"].median())
    df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
    X = df.drop(columns=["PassengerId", "Embarked", "Survived"])
    y = df["Survived"].values
    return X, y
```
