# Chapter 6: Support Vector Machines & Naive Bayes

In this unit, we explore two highly effective classification algorithms, particularly powerful for text mining and boundary separation.

### High-Level Intuition & Analogy
* **Naive Bayes** is based on word frequencies. To classify whether an email is spam, it looks at the probability of each word appearing in spam emails vs. non-spam emails. It is "naive" because it assumes words don't influence each other (e.g. it treats "New" and "York" as completely independent), but it is incredibly fast and works well for text.
* **Support Vector Machine (SVM)** is about drawing borders. Imagine black and white marbles scattered on a table. SVM draws a straight line that separates them while maximizing the empty space (the margin) between the line and the closest marbles (support vectors).

### Core Concepts & Step-by-Step Execution
1. **Naive Bayes Probability Calculation**: Calculates conditional probabilities for each class and chooses the class with the highest probability.
2. **Text Feature Extraction**: Text is converted into numbers using Bag-of-Words or TF-IDF representations.
3. **SVM Margin Maximization**: SVM identifies support vectors (points closest to the boundary) and solves a quadratic programming optimization problem to find the optimal decision boundary.
4. **C and Gamma Hyperparameters**:
   * `C`: Controls the trade-off between maximizing the margin and minimizing classification errors. A small `C` gives a wide margin but allows some errors.
   * `gamma`: In RBF kernel, controls the radius of influence of a single training point. High `gamma` makes the boundary fit closely around individual points, risking overfitting.

### Mathematical Foundations (Simplified)
* **Bayes' Theorem**: Calculates $P(Class | Word) = \frac{P(Word | Class) \cdot P(Class)}{P(Word)}$.
* **Kernel Trick**: When marbles are not separable by a straight line, SVM uses a kernel (like the RBF kernel) to map data into a higher dimension (like lifting them into 3D space) where a flat sheet can easily cut between them.

### Pros and Cons
* **Pros**: Naive Bayes is extremely fast and robust to noise. SVM handles high-dimensional datasets and non-linear boundaries.
* **Cons**: Naive Bayes' independence assumption is often violated. SVM requires careful hyperparameter tuning and is computationally expensive for large datasets.

### Pro-Tips & Gotchas
* **Feature Scaling for SVM**: SVM relies on calculating Euclidean distances between support vectors. If features are on different scales, columns with large values will dominate, making the boundary incorrect. Always scale features before training an SVM!

### Programming Guide & Key Functions
Here is a comprehensive text analysis and SVM training guide:

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

# --- Text Vectorization and Naive Bayes ---
corpus = ["Free cash now", "Meeting schedule update", "Claim free rewards", "Important project details"]
y_text = [1, 0, 1, 0]  # 1 = Spam, 0 = Ham

# Configure TfidfVectorizer
vec = TfidfVectorizer(
    ngram_range=(1, 2),  # Extract both individual words and word pairs
    max_df=0.9,          # Exclude words appearing in more than 90% of documents
    min_df=1             # Include words appearing in at least 1 document
)
X_tfidf = vec.fit_transform(corpus)

# Multinomial Naive Bayes (alpha=1.0 for Laplace smoothing)
nb = MultinomialNB(alpha=1.0).fit(X_tfidf, y_text)

# --- SVM Kernels Configuration ---
X_numeric = [[1.2, 300], [2.1, 400], [1.5, 350], [5.1, 1000]]
y_numeric = [0, 0, 0, 1]

# SVM requires feature scaling!
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_numeric)

# Fitting SVM with non-linear RBF kernel
svm_rbf = SVC(kernel='rbf', C=1.0, gamma='scale').fit(X_scaled, y_numeric)

# Fitting SVM with linear kernel (useful for text features)
svm_lin = SVC(kernel='linear', C=0.5).fit(X_scaled, y_numeric)
```
