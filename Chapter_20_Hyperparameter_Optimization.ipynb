# Chapter 20: Hyperparameter Optimization with Optuna

Tuning models manually is slow. In this unit, we explore automated optimization strategies.

### High-Level Intuition & Analogy
**Optuna** is a framework that automates the search for model hyperparameters. Instead of guessing randomly, it uses **Bayesian Optimization** - constructing a statistical model of the parameter space to guess which settings are likely to yield the best results.

### Core Concepts & Step-by-Step Execution
1. **Define Objective**: Create a function that takes a `trial` object, suggests parameters, trains the model, and returns a score.
2. **Pruning**: Terminate trials early if they are performing poorly.
3. **Bayesian Updates**: Update the probability model after each trial.

### Mathematical Foundations (Simplified)
* **Bayesian Optimization**: Uses past trial results to build a probability model $P(\text{score} | \text{params})$ to choose the next trial parameters by maximizing **Expected Improvement (EI)**.

### Pros and Cons
* **Pros**: Extremely fast compared to grid search, and automates parameter tuning.
* **Cons**: Needs setup code, and can still overfit hyperparameters to the validation set.

### Pro-Tips & Gotchas
* **Validation Sets**: Always tune hyperparameters on a separate validation set or via cross-validation, never on the test set! Tuning on the test set leads to hyperparameter overfitting.

### Programming Guide & Key Functions
Here is a comprehensive hyperparameter optimization study setup:

```python
import optuna
from sklearn.ensemble import RandomForestClassifier
import numpy as np

X_train, y_train = np.random.rand(50, 2), np.random.randint(0, 2, 50)

# --- Define Objective Function ---
def objective(trial):
    # Suggest numerical options
    n_est = trial.suggest_int('n_estimators', 10, 100, step=10)
    depth = trial.suggest_int('max_depth', 2, 10)
    learning_rate = trial.suggest_float('learning_rate', 1e-4, 1e-1, log=True)
    criterion = trial.suggest_categorical('criterion', ['gini', 'entropy'])
    
    model = RandomForestClassifier(n_estimators=n_est, max_depth=depth, criterion=criterion)
    model.fit(X_train, y_train)
    
    return model.score(X_train, y_train)  # Optimize this score

# --- Study Configuration ---
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=5)

print("Best hyperparameter values:", study.best_params)
print("Best score value:", study.best_value)
```

````python
import numpy as np

import optuna
import logging

# Disable optuna logs to avoid clutter
optuna.logging.set_verbosity(optuna.logging.WARNING)


# Simple optimization demo: find x to minimize (x - 4)^2
def objective(trial):
    x = trial.suggest_float("x", -10, 10)
    return (x - 4) ** 2


study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=50)
print("Best Parameter:", study.best_params)
print("Best Value:", study.best_value)
````
