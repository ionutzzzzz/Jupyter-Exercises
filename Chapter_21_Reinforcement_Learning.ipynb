# Chapter 21: Reinforcement Learning (RL)

Reinforcement Learning (RL) is a paradigm of Machine Learning where an agent learns to make decisions by performing actions in an environment to maximize some cumulative reward.

## 1. Markov Decision Processes (MDPs)
An MDP is mathematically defined by a 5-tuple $(S, A, P, R, \gamma)$:
- $S$: A finite set of states.
- $A$: A finite set of actions.
- $P(s' \mid s, a) = \mathbb{P}(S_{t+1}=s' \mid S_t=s, A_t=a)$: Transition probability matrix.
- $R(s, a, s')$: Reward function.
- $\gamma \in [0, 1)$: Discount factor balancing immediate vs. future rewards.

## 2. Q-Learning
Q-Learning is an off-policy temporal-difference algorithm. It estimates the quality of action-state pairs $Q(s, a)$ which represents the expected cumulative discounted reward of taking action $a$ in state $s$ and thereafter following the optimal policy.

The temporal difference (TD) Bellman update is:
$$Q(s, a) \leftarrow Q(s, a) + \alpha \left[ R(s, a, s') + \gamma \max_{a'} Q(s', a') - Q(s, a) \right]$$
Where $\alpha$ is the learning rate.

## 3. Exploration vs. Exploitation
An agent must balance exploring new paths and exploiting known good paths. In the **$\epsilon$-greedy** strategy:
- With probability $\epsilon$, select a random action (exploration).
- With probability $1-\epsilon$, select $\arg\max_a Q(s, a)$ (exploitation).

## 🛠️ Key Functions to Keep in Mind
* **NumPy Functions**:
  * `np.zeros(shape)`: Initialize array of zeros.
  * `np.argmax(arr)`: Find the index of the maximum value.
  * `np.random.RandomState(seed)`: Random number generator.


```python
import numpy as np
import random

# Example of epsilon-greedy logic
q_values = np.array([0.5, 2.0, 1.0, 0.2])
epsilon = 0.2
if random.random() < epsilon:
    action = random.randint(0, len(q_values)-1)
    print(f"Explored random action: {action}")
else:
    action = np.argmax(q_values)
    print(f"Exploited greedy action: {action}")
```
