# Chapter 21: Reinforcement Learning (RL)

Reinforcement Learning trains agents to make decisions by rewarding desired behaviors and punishing undesired ones.

### High-Level Intuition & Analogy
This is learning by trial-and-error (like training a dog). An agent takes actions in an environment, receives rewards, and updates a memory table (**Q-Table**) of action values.

### Core Concepts & Math
1. **Action Selection**: Select action using $\epsilon$-greedy exploration.
2. **Step execution**: Take action, observe reward $R$ and next state $s'$.
3. **Temporal Difference calculation**: Compare current Q-value to target.
4. **Q-Table Update**: Update Q-value based on learning rate.

### Mathematical Foundations (Simplified)
* **Q-Learning Update Rule**:
  $$Q(s, a) \leftarrow Q(s, a) + \alpha \left[ R + \gamma \max_{a'} Q(s', a') - Q(s, a) \right]$$
  Where $\alpha$ is the learning rate, $\gamma$ is the discount factor, and $R$ is the reward.

### Pros and Cons
* **Pros**: Learns complex behaviors without human supervision, and adapts to dynamic environments.
* **Cons**: Highly sensitive to reward functions, and takes millions of training steps to converge.

### Pro-Tips & Gotchas
* **Exploration Decay**: Start with high exploration ($\epsilon = 1.0$) to let the agent explore the environment, and slowly decay it ($\epsilon \leftarrow \epsilon \cdot 0.99$) as training progresses so the agent starts exploiting its learned Q-table.

### Programming Guide & Key Functions
Here is a complete setup code representing discrete action states and update loops:

```python
import numpy as np

# Q-Table environment setup
num_states = 5
num_actions = 2
q_table = np.zeros((num_states, num_actions))

alpha = 0.1    # Learning rate
gamma = 0.9    # Discount factor
epsilon = 0.2  # Exploration threshold

# --- State Transition Update Loop ---
state = 0
action = 1
reward = 1.0
next_state = 1

# Q-learning TD Update logic
td_target = reward + gamma * np.max(q_table[next_state])
q_table[state, action] += alpha * (td_target - q_table[state, action])
```

````python
import numpy as np
import random

# Example of epsilon-greedy logic
q_values = np.array([0.5, 2.0, 1.0, 0.2])
epsilon = 0.2
if random.random() < epsilon:
    action = random.randint(0, len(q_values) - 1)
    print(f"Explored random action: {action}")
else:
    action = np.argmax(q_values)
    print(f"Exploited greedy action: {action}")
````
