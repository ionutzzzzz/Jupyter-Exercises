# Chapter 12: Natural Language Processing (NLP)

Natural Language Processing (NLP) focuses on teaching computers to read, write, and understand human language.

### High-Level Intuition & Analogy
To a computer, a word is just a string of characters. We must turn text into numbers.
* First, we clean text by lowercasing, removing punctuation, and dropping generic words like "the" or "is" (**stopwords**).
* Then, we tokenize sentences into words.
* We represent words as vectors (**Embeddings**) where words with similar meanings (e.g. "king" and "queen") point in similar directions in space.

### Core Concepts & Step-by-Step Execution
1. **Preprocessing**: Text is cleaned, tokenized, and stripped of stopwords.
2. **TF-IDF Calculation**: Scoring words by term frequency multiplied by inverse document frequency.
3. **Similarity Comparison**: Using cosine similarity to compare document vectors.

### Mathematical Foundations (Simplified)
* **TF-IDF Formula**:
  $$\text{TF-IDF}(t, d, D) = \text{TF}(t, d) \cdot \log\left(\frac{N}{1 + \text{DF}(t, D)}\right)$$
  Where $t$ is the word, $d$ is the document, $D$ is the collection of all documents, and $N$ is the total number of documents.
* **Cosine Similarity**: Measures the angle between two text vectors. Formula: $\cos(\theta) = \frac{\mathbf{u} \cdot \mathbf{v}}{\|\mathbf{u}\| \|\mathbf{v}\|}$. If the vectors point in the same direction, similarity is 1.

### Pros and Cons
* **Pros**: Simple text classification, handles large vocabularies, and captures document similarities.
* **Cons**: Bag-of-words ignores word order (e.g. "not bad" and "bad, not" look identical), and cannot handle context.

### Pro-Tips & Gotchas
* **Sparsity**: TF-IDF matrices can contain thousands of columns but only a few non-zero entries per row (high sparsity). Use sparse matrix formats and prune rare words using TfidfVectorizer parameters like `min_df` and `max_df`.

### Programming Guide & Key Functions
Here is a comprehensive text parsing and similarity ranking script:

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Ensure required resources are downloaded
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)

# --- NLTK Text Preprocessing ---
raw_text = "The quick brown fox jumps over the lazy dog."
tokens = word_tokenize(raw_text.lower())  # Tokenize and lowercase

# Filter stopwords
english_stops = set(stopwords.words('english'))
filtered_tokens = [w for w in tokens if w.isalpha() and w not in english_stops]

# Lemmatization (normalize words to base forms, e.g., 'running' -> 'run')
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(t) for t in filtered_tokens]

# --- Cosine Similarity Matrix ---
corpus = ["Machine learning is fun", "Deep learning is also fun", "Cooking food is a nice hobby"]
vec = TfidfVectorizer()
tfidf = vec.fit_transform(corpus)

sim_matrix = cosine_similarity(tfidf)  # Pairwise similarities
print("Cosine similarity between doc 1 and doc 2:", sim_matrix[0, 1])
```

````python
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# TF-IDF Demonstration
docs = ["I love machine learning.", "Machine learning is powerful."]
vec = TfidfVectorizer()
tfidf = vec.fit_transform(docs)

# Output representation
print("Vocabulary:", vec.vocabulary_)
print("TF-IDF Matrix Shape:", tfidf.shape)

# Cosine Similarity
sim = cosine_similarity(tfidf[0], tfidf[1])
print(f"Similarity between document 0 and 1: {sim[0][0]:.4f}")
````
