### Exercise 5: Lasso Regression for Feature Selection (Medium/Hard)
Lasso regression (L1 regularization) can drive coefficients exactly to zero, performing automatic feature selection.
Write a function `lasso_feature_selection(X, y, feature_names)` that:
1. Fits a `sklearn.linear_model.Lasso(alpha=10.0, random_state=42)` model on features `X` and target `y`.
2. Identifies the features that have non-zero coefficients.
3. Returns a list of strings of the names of these selected features.


```python
from sklearn.linear_model import Lasso


def lasso_feature_selection(X, y, feature_names):
    # TODO: Fit Lasso(alpha=10.0, random_state=42) and return list of features with non-zero coefficients
    model = Lasso(alpha=10.0, random_state=42)
    model.fit(X, y)
    non_zero_indices = np.where(model.coef_ != 0)[0]
    return [feature_names[i] for i in non_zero_indices]
```
