# Chapter 23: Graph Machine Learning & Link Analysis

Graph ML studies datasets structured as networks (nodes and edges).

### High-Level Intuition & Analogy
Many real-world datasets are networks of interconnected items (like social media friends or web pages). **Graph ML** studies these relationships, predicting node roles or link connections.

### Core Concepts & Step-by-Step Execution
1. **Representation**: Model networks using adjacency and degree matrices.
2. **Eigenvalue Analysis**: Compute graph Laplacians to study structural properties.
3. **Message Passing**: Aggregate feature representation vectors from neighboring nodes.

### Mathematical Foundations (Simplified)
* **Graph Laplacian**: $L = D - A$.
* **PageRank Equation**:
  $$\mathbf{r} = d \cdot M \cdot \mathbf{r} + \frac{1 - d}{N} \mathbf{1}$$
  Where $d$ is the damping factor, $M$ is the transition probability matrix, and $N$ is the number of nodes.

### Pros and Cons
* **Pros**: Models relationships and structures naturally.
* **Cons**: Computationally expensive, and hard to scale to very large graphs.

### Pro-Tips & Gotchas
* **Over-smoothing**: Stacking too many graph convolutional layers causes all node representation vectors to converge to the same average values, making node classification impossible. Limit model depth to 2-3 layers.

### Programming Guide & Key Functions
Here is a comprehensive script creating adjacency, degree, and Laplacian matrices:

```python
import numpy as np

# Network adjacency structure (3 connected nodes)
# Node 0 connected to 1, Node 1 connected to 0 and 2, Node 2 connected to 1
A = np.array([
    [0, 1, 0],
    [1, 0, 1],
    [0, 1, 0]
])

# --- Calculate Degree Matrix D ---
degrees = np.sum(A, axis=1)  # Sum rows
D = np.diag(degrees)  # Diagonal matrix of degree counts

# --- Calculate Graph Laplacian L ---
L = D - A

# --- PageRank iteration step ---
N = 3
damping = 0.85
r = np.array([1/N, 1/N, 1/N])  # Initial node ranks
transition_matrix = np.array([
    [0,   0.5, 0],
    [1.0, 0,   1.0],
    [0,   0.5, 0]
])
# Single iteration step
r = damping * np.dot(transition_matrix, r) + (1 - damping) / N
```
