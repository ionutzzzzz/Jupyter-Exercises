# Chapter 23: Graph Machine Learning & Link Analysis

Graphs represent entities (nodes) and their relationships (edges). Graph Machine Learning focuses on node classification, link prediction, and learning dense node representations.

## 1. Graph Representation
An unweighted, undirected graph is represented by an $N \times N$ Adjacency Matrix $A$:
- $A_{i, j} = 1$ if there is an edge between node $i$ and node $j$, else $0$.
- Node Degree: The number of connections a node has. The Degree Matrix $D$ is diagonal where $D_{i,i} = \sum_j A_{i,j}$.
- Graph Laplacian: $L = D - A$. It captures diffusion processes and forms the foundation for spectral graph theory.

## 2. Link Prediction & Jaccard Coefficient
Given nodes $u$ and $v$ and their neighbors $\Gamma(u)$ and $\Gamma(v)$:
$$J(u, v) = \frac{|\Gamma(u) \cap \Gamma(v)|}{|\Gamma(u) \cup \Gamma(v)|}$$

## 3. PageRank Algorithm
PageRank computes the importance of nodes recursively. The PageRank score vector $r$ is updated as:
$$r = d \cdot M \cdot r + \frac{1 - d}{N} \mathbf{1}$$
Where $M$ is the column-stochastic transition probability matrix, and $d$ is the damping factor (typically 0.85).

## 4. Graph Embeddings via Random Walks
Algorithms like DeepWalk use random walks to generate node sequences, which are then treated like sentences in Word2Vec to learn low-dimensional feature embeddings.

  * `np.diag(vector)`: Create diagonal matrices.
  * `np.dot(A, B)`: Matrix multiplication.
  * `np.where(condition)`: Indices matching conditions.
