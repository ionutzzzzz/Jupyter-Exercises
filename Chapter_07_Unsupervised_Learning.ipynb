# Chapter 7: Unsupervised Learning & Clustering

Unsupervised learning is used when we have inputs but no pre-existing answers (labels). We want the computer to discover hidden structures on its own.

### High-Level Intuition & Analogy
* **K-Means Clustering** is like sorting scattered objects into $K$ separate piles. We randomly drop $K$ flags (centroids), assign objects to the nearest flag, move each flag to the center of its pile, and repeat until the piles stop moving.
* **PCA (Principal Component Analysis)** is like photographing a 3D object. A camera drops the depth dimension (3D to 2D) but chooses an angle that captures the object's width and height as clearly as possible. PCA finds the directions of highest spread (variance) to compress variables.

### Core Concepts & Step-by-Step Execution
1. **K-Means Iteration**:
   * Initialize $K$ cluster centroids randomly.
   * Assign each data point to the closest centroid.
   * Move each centroid to the mean of all points assigned to it.
   * Repeat until centroids stop shifting.
2. **PCA Computation**:
   * Center the data around its mean.
   * Calculate the covariance matrix showing how features vary together.
   * Find eigenvectors (principal components) and project the data onto them.

### Mathematical Foundations (Simplified)
* **K-Means Objective (Inertia)**: Sum of squared distances of samples to their closest cluster center.
  $$J = \sum_{i=1}^{N} \min_{\mu_j} ||x_i - \mu_j||^2$$
* **Elbow Method**: We plot inertia for different values of $K$. The "elbow" point is where adding more clusters yields diminishing returns in reducing inertia.

### Pros and Cons
* **Pros**: Simple to implement, PCA speeds up modeling, and discovers hidden groups.
* **Cons**: K-Means is highly sensitive to outlier points and initial centroid positions. PCA makes features un-interpretable because they are linear combinations of the originals.
* **Clustering Alternatives**: If clusters have irregular shapes, K-Means will fail (it assumes spherical clusters). Use **DBSCAN** (Density-Based Spatial Clustering) which groups dense neighborhoods and labels sparse areas as noise outliers.

### Pro-Tips & Gotchas
* **Outliers and K-Means**: Because K-Means recalculates centroids using averages, outliers can pull centroids far away from natural clusters. Clean outliers before clustering!

### Programming Guide & Key Functions
Here is a comprehensive guide to clustering and PCA functions:

```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.decomposition import PCA
import numpy as np

X = np.array([[1, 2], [1.5, 1.8], [5, 8], [8, 8], [1, 0.6], [9, 11]])

# --- K-Means Configuration ---
kmeans = KMeans(
    n_clusters=2,
    init='k-means++',   # Smart initialization to prevent poor local minimums
    n_init=10,          # Number of times K-Means runs with different random centroids
    random_state=42
)
kmeans.fit(X)
print("Centroids:", kmeans.cluster_centers_)
print("Inertia:", kmeans.inertia_)

# --- DBSCAN Density-Based Clustering ---
dbscan = DBSCAN(eps=1.5, min_samples=2)  # Radius threshold eps, min density samples
labels = dbscan.fit_predict(X)  # Outliers are assigned label -1

# --- PCA Compression ---
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

print("Explained variance ratio:", pca.explained_variance_ratio_)  # [Var_PC1, Var_PC2]
print("Components directions:", pca.components_)
```

````python
import numpy as np

from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load data
iris = load_iris()
X = iris.data

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
print("Original shape:", X.shape)
print("PCA shape:", X_pca.shape)
````
