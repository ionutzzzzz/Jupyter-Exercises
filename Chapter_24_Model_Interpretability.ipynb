# Chapter 24: Model Interpretability & Explainable AI (XAI)

XAI focuses on explaining the predictions of complex models.

### High-Level Intuition & Analogy
If a model makes predictions, we need to understand why.
* **Permutation Importance** shuffles a feature's values. If accuracy drops significantly, that feature is important.
* **SHAP Values** use cooperative game theory to measure the contribution of each feature to a prediction.

### Core Concepts & Step-by-Step Execution
1. **Permutation Shuffle**: Record baseline accuracy, shuffle a feature column, recalculate accuracy, and measure the drop.
2. **SHAP Allocation**: Estimate contribution of features by calculating predictions across all feature coalitions.

### Mathematical Foundations (Simplified)
* **Shapley Value**:
  $$\phi_i(v) = \sum_{S \subseteq N \setminus \{i\}} \frac{|S|!(N - |S| - 1)!}{N!} (v(S \cup \{i\}) - v(S))$$
  We calculate the marginal contribution of feature $i$ across all possible feature subsets $S$.

### Pros and Cons
* **Pros**: Builds trust, aids model debugging, and identifies bias.
* **Cons**: Explanations are approximations, and SHAP is slow to calculate.

### Pro-Tips & Gotchas
* **Correlation Trap**: Shuffling a highly correlated feature can create unrealistic samples (e.g. `Age=5` but `YearsOfWork=30`), causing misleading drops in accuracy during permutation importance.

### Programming Guide & Key Functions
Here is a comprehensive script using Scikit-Learn to evaluate permutation importances:

```python
from sklearn.inspection import permutation_importance
from sklearn.linear_model import LogisticRegression
import numpy as np

# Sample training
X = np.array([[1, 10], [2, 20], [3, 30], [4, 40], [5, 50]])
y = np.array([0, 0, 1, 1, 1])

model = LogisticRegression().fit(X, y)

# --- Permutation Feature Importance ---
result = permutation_importance(
    model, 
    X, 
    y, 
    n_repeats=5,       # Number of times to shuffle each feature column
    random_state=42, 
    scoring='accuracy'
)

# Extract statistics
mean_importance = result.importances_mean  # [Importance_Col0, Importance_Col1]
std_importance = result.importances_std    # Standard deviations across shuffles
```

````python
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.inspection import permutation_importance

# Example of linear model coefficient feature importance
X = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])
y = np.array([2.5, 4.5, 6.5, 8.5])
model = Ridge().fit(X, y)
print("Model Coefficients:", model.coef_)
# Feature importance scaled by standard deviations
importances = np.abs(model.coef_) * np.std(X, axis=0)
print("Scaled Importances:", importances)
````
