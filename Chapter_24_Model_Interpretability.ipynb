# Chapter 24: Model Interpretability & Explainable AI (XAI)

Explainable AI (XAI) refers to methods and techniques in the application of artificial intelligence technology such that the results of the solution can be understood by human experts.

## 1. Permutation Feature Importance
Permutation feature importance measures the increase in the prediction error of the model after we shuffle the values of a feature, which breaks the relationship between the feature and the target.
- If shuffling a feature's values increases model error significantly, the feature is highly important.
- If shuffling has no effect, the feature is not important.

## 2. Partial Dependence Plots (PDP)
Partial Dependence Plots show the marginal effect of one or two features on the predicted outcome of a machine learning model. It isolates the feature and averages the model predictions over all samples:
$$\hat{f}_{x_S}(x_S) = \frac{1}{n} \sum_{i=1}^n f(x_S, x_C^{(i)})$$
Where $x_S$ is the target feature and $x_C$ are all other features.

## 3. Shapley Values (SHAP)
Shapley values originate from cooperative game theory. They attribute a fair share of the prediction to each feature by computing the average marginal contribution of the feature across all possible feature subsets (coalitions) $S \subseteq F \setminus \{i\}$:
$$\phi_i = \sum_{S \subseteq F \setminus \{i\}} \frac{|S|!(|F| - |S| - 1)!}{|F|!} \left[ v(S \cup \{i\}) - v(S) \right]$$
Where $F$ is the set of all features, and $v(S)$ is the characteristic value function (model prediction with subset $S$).

## 🛠️ Key Functions to Keep in Mind
* **Scikit-learn Interpretability**:
  * `permutation_importance(model, X, y, random_state)`: Scikit-learn permutation importance.
* **NumPy Functions**:
  * `np.mean(arr, axis)`: Compute array mean.
  * `rng.shuffle(arr)`: In-place shuffle (using `np.random.RandomState`).


```python
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.inspection import permutation_importance

# Example of linear model coefficient feature importance
X = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])
y = np.array([2.5, 4.5, 6.5, 8.5])
model = Ridge().fit(X, y)
print("Model Coefficients:", model.coef_)
# Feature importance scaled by standard deviations
importances = np.abs(model.coef_) * np.std(X, axis=0)
print("Scaled Importances:", importances)
```
