# Chapter 14: Ensemble Learning & Gradient Boosting

Ensemble learning combines predictions from multiple individual models to create a stronger, more robust model.

### High-Level Intuition & Analogy
* **Voting Classifiers** let different models (e.g. Logistic Regression, SVM, Decision Tree) vote on the prediction. Hard voting uses majority rule, while soft voting averages the predicted probabilities.
* **Boosting** trains models sequentially. The first model makes predictions, the second model is trained specifically on the errors (residuals) of the first, the third on the errors of the second, and so on.

### Core Concepts & Step-by-Step Execution
1. **Voting Ensemble**: Combine diverse base classifiers to reduce variance.
2. **AdaBoost Iteration**:
   * Assign equal weight to all training samples.
   * Train a weak model (e.g. a shallow tree).
   * Increase weight of misclassified samples.
   * Train next model.
3. **Gradient Boosting**: Compute residuals (errors), train a tree to predict residuals, add predictions to the model, and repeat.

### Mathematical Foundations (Simplified)
* **Soft Voting Formula**:
  $$\hat{p} = \frac{1}{M} \sum_{m=1}^{M} p_m$$
  Where $M$ is the number of models and $p_m$ is the probability predicted by model $m$.

### Pros and Cons
* **Pros**: Extremely high predictive accuracy, and robust to overfitting.
* **Cons**: High computational cost, and acts as a "black box" model.

### Pro-Tips & Gotchas
* **Learning Rate vs. Estimators**: In Gradient Boosting, there is a trade-off between the `learning_rate` and `n_estimators`. A smaller learning rate requires more estimators, which takes longer to train but generally yields better generalization performance.

### Programming Guide & Key Functions
Here is a comprehensive ensemble structures script:

```python
from sklearn.ensemble import VotingClassifier, AdaBoostClassifier, GradientBoostingClassifier, StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

X_train, y_train = [[1], [2], [3], [4], [5]], [0, 0, 0, 1, 1]

# --- Soft Voting Ensemble Configuration ---
clf1 = LogisticRegression()
clf2 = SVC(probability=True)
voting = VotingClassifier(
    estimators=[('lr', clf1), ('svm', clf2)],
    voting='soft',
    weights=[1, 2]  # Assign relative weight influence to models
)
voting.fit(X_train, y_train)

# --- Boosting Classifiers ---
ada = AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1), n_estimators=50)
ada.fit(X_train, y_train)

gb = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3)
gb.fit(X_train, y_train)

# --- Stacking Classifier ---
# Stacking trains a meta-model on predictions of base estimators
stack = StackingClassifier(
    estimators=[('lr', clf1), ('dt', DecisionTreeClassifier(max_depth=2))],
    final_estimator=LogisticRegression()
)
stack.fit(X_train, y_train)
```

````python
import numpy as np

from sklearn.ensemble import VotingClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# Base models
clf1 = LogisticRegression(random_state=42)
clf2 = DecisionTreeClassifier(max_depth=3, random_state=42)

# Voting ensemble
voting = VotingClassifier(estimators=[("lr", clf1), ("dt", clf2)], voting="soft")
X, y = make_classification(n_samples=100, random_state=42)
voting.fit(X, y)
print("Voting Score:", voting.score(X, y))
````
