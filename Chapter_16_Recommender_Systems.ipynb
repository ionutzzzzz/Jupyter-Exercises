# Chapter 16: Recommender Systems

Recommender systems recommend items (like books or movies) to users based on preferences.

### High-Level Intuition & Analogy
* **Content-Based**: Recommending items similar to what the user liked before (e.g. recommending space movies to someone who watched *Interstellar*).
* **Collaborative Filtering**: Finding users with similar tastes and recommending items they liked.
* **Matrix Factorization**: Decomposes a large, sparse user-item rating table into user-preference and item-attribute factors using Stochastic Gradient Descent (SGD).

### Core Concepts & Step-by-Step Execution
1. **Profile Building**: Create representation vectors for users and items.
2. **Neighborhood Evaluation**: Calculate similarities between users or items.
3. **Factorization Optimization**: Decompose rating matrix $R \approx U \cdot V^T$ and update latent factors using SGD.

### Mathematical Foundations (Simplified)
* **Cosine Similarity**: Formula: $\cos(\theta) = \frac{\mathbf{u} \cdot \mathbf{v}}{\|\mathbf{u}\| \|\mathbf{v}\|}$.
* **Latent Factor Prediction**: Predicted rating $r_{u,i} \approx \mathbf{p}_u^T \mathbf{q}_i$. Update step: $\mathbf{p}_u \leftarrow \mathbf{p}_u + \gamma (e_{u,i} \mathbf{q}_i - \lambda \mathbf{p}_u)$, where $e_{u,i}$ is rating error.

### Pros and Cons
* **Pros**: Personalizes user experiences, and increases sales/engagement.
* **Cons**: The **Cold Start Problem** (cannot recommend items to new users with no history).

### Pro-Tips & Gotchas
* **Sparsity**: Recommender ratings matrices are usually $>99\%$ empty. Neighborhood methods fail on highly sparse matrices; use Matrix Factorization or Deep Learning recommenders instead.

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import TruncatedSVD
import numpy as np

# Dense user-item rating matrix
R = np.array([
    [5, 3, 0, 1],
    [4, 0, 0, 1],
    [1, 1, 0, 5],
    [1, 0, 0, 4],
    [0, 1, 5, 4]
])

# --- Cosine Similarity for Item-Item Neighborhoods ---
item_similarity = cosine_similarity(R.T)
print("Similarity matrix between items:
", item_similarity)

# --- Truncated SVD Matrix Factorization ---
svd = TruncatedSVD(n_components=2, algorithm='randomized', random_state=42)
latent_user_features = svd.fit_transform(R)
latent_item_features = svd.components_

print("Latent user representation shape:", latent_user_features.shape)
print("Latent item representation shape:", latent_item_features.shape)
```
