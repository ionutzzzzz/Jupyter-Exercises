# Chapter 11: Time Series Analysis & Forecasting

Time series data consists of observations ordered sequentially in time.

### High-Level Intuition & Analogy
Imagine predicting tomorrow's temperature. It is highly likely to be similar to today's temperature (autocorrelation), and it follows seasonal cycles. Forecasting requires analyzing trends (long-term directions) and seasonal patterns. We must ensure the series is "stationary" - meaning its average value and variance don't drift over time - before fitting forecasting models.

### Core Concepts & Step-by-Step Execution
1. **Rolling Statistics**: Calculating average and standard deviation over sliding windows to analyze trends.
2. **Time Series Testing**: Using the Augmented Dickey-Fuller (ADF) test.
3. **Differencing**: Subtracting yesterday's value from today's value ($y'_t = y_t - y_{t-1}$) to remove linear trends and make the series stationary.
4. **ARIMA Modeling**: Fitting Autoregressive, Integrated, and Moving Average components to capture temporal correlations.

### Mathematical Foundations (Simplified)
* **ADF Test**: Checks the null hypothesis that a unit root is present (non-stationary). If the p-value is $< 0.05$, we reject the null hypothesis and conclude the series is stationary.
* **ARIMA parameters**:
  * **AR (p)**: Predicts future values based on past values.
  * **I (d)**: Number of differencing steps needed to make the data stationary.
  * **MA (q)**: Predicts future values based on past forecast errors.

### Pros and Cons
* **Pros**: Captures seasonal and trend variations, and works well on sequential data.
* **Cons**: Assumes history repeats, and cannot predict black-swan events.

### Pro-Tips & Gotchas
* **Chronological Splitting**: Never use random splits (like standard train_test_split) for time series data! If you randomly split, you will predict past values using future values, which is data leakage. Always use a chronological split (e.g. train on the first 80% of dates, test on the remaining 20%).

### Programming Guide & Key Functions
Here is a comprehensive time series decomposition and forecasting guide:

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Generate sample sales data
dates = pd.date_range(start='2026-01-01', periods=10, freq='D')
sales = [100, 105, 102, 110, 108, 115, 112, 120, 118, 125]
df = pd.DataFrame({'Sales': sales}, index=dates)

# --- Time Series Decomposition ---
decomposition = seasonal_decompose(df['Sales'], model='additive', period=2)
# Access decomposed components via decomposition.trend, decomposition.seasonal, decomposition.resid

# --- Stationarity Diagnostic ---
result = adfuller(df['Sales'])
p_value = result[1]
print("ADF test statistic:", result[0])
print("p-value:", p_value)

# Make series stationary using differencing
df_diff = df['Sales'].diff().dropna()

# --- ARIMA Forecast Training ---
model = ARIMA(df['Sales'], order=(1, 1, 1))  # ARIMA(p, d, q) parameters
fitted_model = model.fit()

forecast = fitted_model.forecast(steps=3)  # Predict next 3 days
```

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA

# Generate random walk with trend
np.random.seed(42)
time = pd.date_range(start="2026-01-01", periods=100, freq="D")
value = (
    np.sin(np.linspace(0, 20, 100))
    + np.random.normal(scale=0.1, size=100)
    + np.linspace(0, 5, 100)
)
df = pd.DataFrame({"value": value}, index=time)

# Visualizing Rolling Statistics
rolling_mean = df["value"].rolling(window=7).mean()
rolling_std = df["value"].rolling(window=7).std()

plt.figure(figsize=(10, 5))
plt.plot(df["value"], label="Original")
plt.plot(rolling_mean, label="Rolling Mean (7d)", color="red")
plt.plot(rolling_std, label="Rolling Std (7d)", color="black")
plt.legend()
plt.title("Rolling Statistics Visualization")
plt.show()

# Perform ADF Test
result = adfuller(df["value"])
print(f"ADF Statistic: {result[0]:.4f}")
print(f"p-value: {result[1]:.4f}")
````
