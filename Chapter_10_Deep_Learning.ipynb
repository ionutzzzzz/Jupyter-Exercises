# Chapter 10: Introduction to Deep Learning

Deep Learning uses multilayered neural networks to solve complex regression and classification tasks.

### High-Level Intuition & Analogy
Deep learning is modeled after biological brains. 
* A **neuron (Perceptron)** takes multiple inputs, multiplies them by weights (importance), sums them up, and decides whether to send a signal forward using an activation function.
* A **Multi-Layer Perceptron (MLP)** stacks these neurons in layers.
* **Backpropagation** is the way neural networks learn: the model makes a guess (forward pass), calculates how wrong it was (loss), and propagates the error backward through the layers to adjust weights.

### Core Concepts & Step-by-Step Execution
1. **Forward Pass**: Data is passed through the network, layer by layer, multiplying by weights, adding biases, and applying activation functions.
2. **Loss Computation**: The output layer makes predictions, and we calculate error using a loss function (like Mean Squared Error).
3. **Backward Pass (Backpropagation)**: Error gradients are computed using the chain rule of calculus, flowing backward from the output layer to the input layer.
4. **Weights Update**: Optimizers (like SGD or Adam) update weights in the opposite direction of the gradients.

### Mathematical Foundations (Simplified)
* **Perceptron Formula**: $y = f(\sum w_i x_i + b)$ where $f$ is an activation function (like ReLU ($max(0,x)$) or Sigmoid) that introduces non-linearity.
* **Backpropagation**: Gradient descent updates: $w \leftarrow w - \alpha \frac{\partial L}{\partial w}$ (where $\alpha$ is the learning rate).

### Pros and Cons
* **Pros**: Learns complex non-linear structures automatically, and scales well with large datasets.
* **Cons**: Prone to overfitting, requires massive computing resources, and behaves as a complete "black box".

### Pro-Tips & Gotchas
* **Vanishing Gradients**: Deep networks can suffer from vanishing gradients if Sigmoid or Tanh activations are stacked too deep, causing weights near the input layer to stop updating. Always prefer **ReLU** activations in hidden layers.

### Programming Guide & Key Functions
Here is a comprehensive guide to building neural networks with scikit-learn:

```python
from sklearn.neural_network import MLPClassifier

X = [[0, 0], [0, 1], [1, 0], [1, 1]]
y = [0, 1, 1, 0]  # XOR logic gate (requires non-linear hidden layers to solve)

# --- Multi-Layer Perceptron Configuration ---
mlp = MLPClassifier(
    hidden_layer_sizes=(10, 5),  # Stacking two hidden layers (10 neurons, then 5 neurons)
    activation='relu',           # Rectified Linear Unit activation function
    solver='adam',               # Fast gradient optimizer
    alpha=1e-4,                  # L2 penalty parameter to prevent weight bloating
    learning_rate_init=0.01,     # Learning step scale
    max_iter=1000,               # Upper limit on epochs
    random_state=42
)

mlp.fit(X, y)
print("Loss curve history:", mlp.loss_curve_)  # Plot this list to check convergence
```

````python
from sklearn.neural_network import MLPClassifier
import numpy as np

# Simple MLP model
X = np.array([[0.0, 0.0], [1.0, 1.0]])
y = np.array([0, 1])

mlp = MLPClassifier(hidden_layer_sizes=(4,), max_iter=1000, random_state=42)
mlp.fit(X, y)
print("Predictions:", mlp.predict(X))
````
